```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
# Cell 2: Load and Explore the Dataset
try:
    # Load Iris dataset
    iris = load_iris()
    df = pd.DataFrame(iris.data, columns=iris.feature_names)
    df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

    # Display first few rows
    display(df.head())

    # Show data types and missing values
    display(df.info())

    # Check for missing values
    print("Missing values per column:")
    print(df.isnull().sum())

except FileNotFoundError:
    print("Error: Dataset file not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
# Cell 3: Basic Data Analysis
try:
    # Basic statistics
    display(df.describe())

    # Group by species
    grouped = df.groupby('species').mean()
    display(grouped)

    print("\nInsight: Virginica has the largest petal dimensions; Setosa the smallest.")
except Exception as e:
    print(f"Analysis error: {e}")
```

```python
# Cell 4: Line Chart (Index vs Sepal Length)
try:
    df['index'] = df.index
    plt.figure(figsize=(8, 4))
    sns.lineplot(x='index', y='sepal length (cm)', data=df)
    plt.title('Line Chart: Sepal Length Over Index')
    plt.xlabel('Index')
    plt.ylabel('Sepal Length (cm)')
    plt.show()
except Exception as e:
    print(e)
```

```python
# Cell 5: Bar Chart - Avg Petal Length per Species
plt.figure(figsize=(6, 4))
sns.barplot(x='species', y='petal length (cm)', data=df)
plt.title('Average Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
# Cell 6: Histogram of Sepal Width
plt.figure(figsize=(6, 4))
sns.histplot(df['sepal width (cm)'], bins=20, kde=True)
plt.title('Sepal Width Distribution')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Cell 7: Scatter Plot - Sepal Length vs Petal Length
plt.figure(figsize=(6, 4))
sns.scatterplot(
    x='sepal length (cm)',
    y='petal length (cm)',
    hue='species',
    data=df
)
plt.title('Sepal Length vs. Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend(title='Species')
plt.show()
```
